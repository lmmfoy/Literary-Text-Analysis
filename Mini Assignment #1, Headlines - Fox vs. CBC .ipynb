```python
import urllib.request
response = urllib.request.urlopen('https://www.foxnews.com/')
FOXhtml = response.read()
```

```python
from bs4 import BeautifulSoup
FOXsoup = BeautifulSoup(FOXhtml, 'lxml')
```

```python
for FOXheadline in FOXsoup.find_all("h2", class_="title"):
    print(FOXheadline.text)
```

```python
CBCresponse = urllib.request.urlopen('https://www.cbc.ca/')
CBChtml = CBCresponse.read()
```

```python
CBCsoup = BeautifulSoup(CBChtml, 'lxml')
```

```python
for CBCheadline in CBCsoup.find_all("h3", class_="headline"):
    print(CBCheadline.text)
```
