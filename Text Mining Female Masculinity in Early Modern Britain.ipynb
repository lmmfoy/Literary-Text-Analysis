## Text Mining Female Masculinity in Sixteenth and Seventeenth-Century Britain

The idea that gender bending is a new phenomenon may be widespread in popular consciousness, but the work of historians in gender and queer history reveals examples of the permeability of gender boundaries throughout human existence. Indeed, a number of historians have suggested that the rigid boundary between men and women we associate with history throughout the ages really coalesced around the end of the eighteenth century.<sup>1</sup> Certainly in the early to mid-eighteenth century – the period I have been focusing my research on – there are a number of examples of gender fluidity. I have been researching female masculinity during this period and discovering a number of “passing men” who dressed and lived as men for days to decades along with many references to “masculine” or “manly” women.  Attitudes toward these individuals were mixed. Though the discovery of a “counterfeit man” was sometimes met with disgust, Dianne Dugaw has written extensively about the laudatory balladry of the mid-seventeenth century onward that praised the exploits of female soldiers and sailors.<sup>2</sup> Women with “manly” personalities were often criticized but, like the heroines of balladry, sometimes praised. I was interested to know whether the attitudes I have observed in eighteenth-century texts are much changed from those of the sixteenth and seventeenth centuries. To that end, I assembled a corpus of texts from this time period, in hopes of gaining greater insight into early modern gender transgression. I wished to explore the portrayals of masculine women in the corpus and determine whether the mixture of condemnation and praise heaped upon these individuals in the eighteenth-century could be found in earlier centuries as well. I explored the collocates of words associated with female masculinity, and performed sentiment analysis on passages of text. 

The terms I chose to explore in this project focused on early modern understanding of masculine women as those who showed unusual willfulness, bravery, or violence. Women who pushed back against male power or showed "manly" courage were variously called viragos, termagants, shrews, and scolds among other terms. 

My hypothesis was that attitudes toward these women would be mixed as they were in later years. I expected women described as "scolds" and the like to be largely disparaged, but I was unsure how "manly women" and "viragos" would be treated.  

     
### Corpus Creation

I assembled my corpus using selections from the VEP EEBO-TCP Collection. The Text Creation Partnership (TCP) is a partnership between ProQuest, the Council on Library and Information Resources, and the Universities of Michigan and Oxford which has worked to create standardized XML/SGML encoded text editions of books from Early English Books Online (EEBO), Eighteenth Century Collections Online (ECCO), and Evans Early American Imprints. Though these collections use optical character recognition (OCR) to make their texts searchable, the resulting output is not very accurate as degradation of the printed material over time (making text blurry or obscured) combined with variance in letter characters and spelling of early modern printers make character recognition difficult. In order to provide accurate electronic texts, TCP hires people to manually key in texts from these collections, a time-intensive but significantly more reliable method than OCR. The project is far from complete, but has so far made about 8,000 texts from ECCO and Evans freely available, along with about 25,000 freely available EEBO texts (along with 35,000 texts available solely to EEBO-TCP partners). I used the freely available texts from EEBO-TCP. This is a collection of English language books and broadsides from the British Isles, British North America, and Europe printed from 1473 to 1700.<sup>3</sup>
   
Even with the accurate transcriptions, text mining the EEBO-TCP corpus is difficult, because before the mid-seventeenth century, there was little standardization of spelling. This means that the same word could be spelled many ways even in texts from the same year. There were also differences in letter use between the beginning and end of the period covered. For example, “y” was often used in place of “i", “v” and “u” were used for both sounds (with “v” being used at the beginning of the word and “u” in the middle of it), and so on.<sup>4</sup> The Visualizing English Print (VEP) project, run by members from the Universities of Wisconsin-Madison and Strathclyde and the Folger Shakespeare Library, provides the freely available EEBO-TCP texts with programmatic standardized spelling.<sup>5</sup> Using VARD, software designed by Alistair Baron of Lancaster University that allows users to standardize spelling variation in historical corpora, VEP modernized much of the spelling of the EEBO-TCP corpus.<sup>6</sup> There are a number of limitations with this approach that the VEP lists on their website and which result in some mistakes in the texts,<sup>7</sup> but the result is a corpus considerably more conducive to text mining. The VEP provides this corpus in the form of plain text files, which I downloaded to build my corpus.

I narrowed my corpus from the 25,368 texts downloaded from VEP to 1,019 texts. These texts made up 100,292,197 tokens, 82,703,346 words, and 1,060,594 unique word types (see Appendix A.1). First, I removed texts from before 1540 in order to take out those likely to contain the highest number of variant spellings, and because I realized that setting more limits on my corpus was likely a good idea. Next I chose my texts by searching within them for a number of key words or phrases relating to early modern female masculinity.<sup>8</sup> A potential issue with the corpus is that it is skewed toward the later decades. I found significantly fewer texts in the decades before 1600. Whether this was because of spelling variance resulting in my search terms missing documents, because I missed terms that were more common in earlier decades, because there are fewer references to masculine women in these earlier texts, or because the EEBO-TCP corpus contains fewer documents from earlier decades, I do not know. 

Though standardization work had been done on the VEP corpus, a number of variant and archaic spellings remained. I reached out to Dr. Baron to obtain a copy of VARD 2, the latest version of the VARD software; however, after spending some time experimenting with it, I decided that I did not have the time needed to master the software and train it on a sample of my corpus (the tool does offer automatic text normalization but Dr. Baron warned that it is significantly less accurate than training the tool before using its semi-automatic function). Instead, I used Python to remove at least some of the irregular spellings (see Appendix B). The remaining corpus still has a number of spelling variants but is noticeably cleaner than what I began with.

The plain text files were stripped of metadata, but information about the documents was provided by VEP in a spreadsheet. I wanted to be able to manipulate the texts by date, so I made a function to find the publishing date of each text in the spreadsheet and then appended the dates to the beginning of each filename (see Appendix A.2).

I then created three functions. The first, concordances_to_tokens turns concordance rows into tokens, the second, get_concordances_by_column, allows the user to filter concordances by column, and the third, as the name suggests, allows for filtering the concordances by year range.

```python
def concordances_to_tokens(concordances, stopwords):
    # loop through each concordance dictionary (i.e. row in the csv file), and grab the concordance
    # join them together with a new line character
    txt = "\n\n".join([concordance['concordance'] for concordance in concordances])
    tokens = nltk.word_tokenize(txt)
    filtered_tokens = [word.lower() for word in tokens if word[0].isalpha() and word.lower() not in stopwords]
    return filtered_tokens
```

```python
#filter list of passages by a given column (e.g. by term with a value of "wife")
def get_concordances_by_column(column, value, passages):
    filteredPassages = [passage for passage in passages if passage[column] == value]
    return filteredPassages    
```

```python
def get_concordances_by_year_range(yearMin, yearMax, passages):
    filteredPassages = []
    for passage in passages:
        # Most lines have one publishing year given, but some have a year range, e.g. 1571-1595
        # I split by the "-" character, to make a list of years, and take the average
        # this will also work with single year values.
        # solution found at: https://www.geeksforgeeks.org/find-average-list-python/
        yearVal = passage["year"]
        years = yearVal.split("-") # this splits the first and second dates in the range, turning the string to an array
        avgYear = sum([int(year) for year in years])/len(years) #strings are turned into integers, added, then divided by the number of years given (i.e. 1 or 2)
        
        if avgYear >= yearMin and avgYear < yearMax: # if the average year is between the minimum and maximum years given
            filteredPassages.append(passage)
    return filteredPassages
```

## Exploring Collocates

### Collocates of normative women terms

In order to find the collocates of words relating to women, I created a function that loops through my tokens and creates a list of collocates. I made two lists of words relating to normative femininity which I wanted to find collocates for. The first list consisted of nouns, the second of adjectives.

```python
lady_words_1 = ["female", "females", "wife", "wives", "lady", "ladies", "woman", "women", "bride", "brides", "damsel", "damsels", "daughter", "daughters", "widow", "widows", "gentlewoman", "gentlewomen", "girl", "girls", "lass", "lasses", "maiden", "maidens", "mother", "mothers", "sister", "sisters", "womanhood", "niece", "nieces", "virgin", "virgins", "matron", "matrons"]

lady_words_2 = ["feminine", "womanly", "girlish", "ladylike", "maidenly", "matronly", "motherly", "maternal", "maternally"]
```

I created a function to find the collocates of a word in the list of tokens given:

```python
# code adapted from: https://nbviewer.jupyter.org/github/sgsinclair/alta/blob/master/ipynb/utilities/Collocates.ipynb

def word_to_find_function(word_to_find, tokens):
    context = 5 # Sets the context of words on either side to grab

    end = len(tokens)
    counter = 0
    the_collocates = []
    for word in tokens:
        if word == word_to_find: # This checks to see if the word is what we want
            for i in range(context):
                if (counter - (i + 1)) >= 0: # Checks that we aren't at the beginning
                    the_collocates.append(tokens[(counter - (i + 1))]) # Adds words before
                if (counter + (i + 1)) < end: # Checks that we aren't at the end
                    the_collocates.append(tokens[(counter + (i + 1))]) # Adds words after
        counter = counter + 1
    
    return(the_collocates)
```

Using the word_to_find_function and matplotlib.pyplot, I then looped through my tokens in the normative women text sample to find collocates to the words in my lady_words_1 and lady_words_2 lists and graphed them.

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(lady_words_grid_rows, 2, figsize=(20, 40)) # define the number of row, columns, and overall size
# These dimensions are not ideal, but after trying many different sizes, this was most readable

for idx, word in enumerate(lady_words_1):
    collocate = word_to_find_function(word, fem_tokens)
    freqs = nltk.FreqDist(collocate).most_common(20)
    plt.subplot(lady_words_grid_rows, 2, (idx+1)) # select the grid location that corresponds to the index
    plt.plot([count for token, count in freqs]) # plot the counts
    plt.xticks(range(len(freqs)), [token for token, count in freqs], rotation=45) # plot the words
    plt.title(word) # set the title
    plt.subplots_adjust(hspace=2) # adjusting the space found between graphs. Solution found here: https://stackoverflow.com/questions/41007675/python-matplotlib-figsize-for-subplots-adding-space-between-rows
```

The collocates show some surprising and some unsurprising connections. Many women terms are heavily collocated with "men," "fathers," "sons," and so on. This likely reflects references to "women and men," and the tendency to find different people discussed in the same passage, but perhaps also points to a tendency on the part of early modern writers to discuss women in terms of their relationship to men. It would be interesting in another study to compare these results against words found as collocates to male terms. 

Women terms are also frequently collocated with "God," which is perhaps to be expected as God and religion are a frequent topic for these texts. Less expected, is that these terms also collocate frequently with "war." Whether this is because of the way the texts for my corpus were chosen (they might include descriptions of manly women going to war, etc.), because war was a common topic and collocates with both genders heavily, or through a quirk in my concordances is unclear. 

The words used to describe women in these concordances seem to be largely positive ones. "Female" is correlated to "kind" and "good," "lady" to "lover," "fair," and "good," "matron" with "chaste" and "noble," "girls" with "merry," and so on. Other descriptive words that collocate are great, old, young, poor, little, and beloved. 

Another positive word that makes a frequent appearance is "love." 

There are other interesting results. References to "maidens" appear to collocate strongly with references to breasts: "breasts," "round," "dugs," "swell." When a "virgin" is refered to she seems frequently to be the Virgin Mary, and "virgin" often seems to refer to the vestal virgins. "Matrons" meanwhile appear to be discussed in the context of (probably historical) Rome. References to "brides" and "bride" collocate with terms referring to the wedding night: "chamber," "night," etc.

Finally, the collocates of "womanhood" point to the normative values of femininity during this era: "chastity,"shamefastness," and "modesty"

I was curious to see what the collocates of the entire lists of normative female words would be, so I made a function that adapted the original word_to_find_function to be able to take word list and token list variables.

```python
def word_list_collocates(word_list, tokens):
    
    context = 5 # This sets the context of words on either side to grab

    end = len(tokens)
    collocates = []
    counter = 0
    for token in tokens:
        if token in word_list: # This checks to see if the word is what we want
            for i in range(context):
                if (counter - (i + 1)) >= 0: # This checks that we aren't at the beginning
                    collocates.append(tokens[(counter - (i + 1))]) # This adds words before
                if (counter + (i + 1)) < end: # This checks that we aren't at the end
                    collocates.append(tokens[(counter + (i + 1))]) # This adds words after
        counter = counter + 1

    return collocates
```

I then ran my lists of normative lady words through them.

```python
lady_words_1_collocates = word_list_collocates(lady_words_1, fem_tokens)
```

```python
lady_words_2_collocates = word_list_collocates(lady_words_2, fem_tokens)
```

## Sentiment Analysis

My next step was to use sentiment analysis to output sentiment scores for relevant words.

```python
from nltk.corpus import sentiwordnet as swn
```

This function is used to map and filter the part-of-speech tags from nltk.pos_tag to synset parts of speech.

```python
def treebank_to_wordnet_pos(treebank, skipWordNetPos=[]):
    if "NN" in treebank and "n" not in skipWordNetPos: # singular and plural nouns (NN, NNS)
        return "n"
    elif "JJ" in treebank and "a" not in skipWordNetPos: # adjectives including comparatives and superlatives (JJ, JJR, JJS)
        return "a" 
    elif "VB" in treebank and "v" not in skipWordNetPos: # verbs in various forms (VB, VBD, VBG, VBN, VBP, VBZ)
        return "v"
    elif "RB" in treebank and "r" not in skipWordNetPos: # adverbs including comparatives and superlatives (RB, RBR, RBS)
        return "r"
    # if we don't match any of these we implicitly return None
```

This function takes a token and returns the sentiment score:

```python
def get_sentiment_score_from_tagged(token, treebank, skipWordNetPos=[]):
    wordnet_pos = treebank_to_wordnet_pos(treebank, skipWordNetPos)
    if wordnet_pos: # only print matches
        senti_synsets = list(swn.senti_synsets(token, wordnet_pos)) # calls the list of synsets and their sentiment scores
        if senti_synsets:
            return senti_synsets[0].pos_score() - senti_synsets[0].neg_score() # subtract the negative score from the positive score
```

This function loops through the tagged tokens, keeps track of positive and negative words, and tracks the total sentiment score.

```python
def get_sentiment_data_from_tokens(tokens, skipWordNetPos=[]):
    tagged = nltk.pos_tag(tokens) # tags parts of speech
    positives = []
    negatives = []
    tokens_score = 0
    for word, treebank in tagged:
        score = get_sentiment_score_from_tagged(word, treebank, skipWordNetPos) # call previous function
        if score:
            tokens_score += score # keeps tally of score 
            if score > 0:
                positives.append(word.lower()) # tallies positive words
            else:
                negatives.append(word.lower()) # tallies negative words
    return tokens_score, positives, negatives
```

I wanted to be able to compare across decades to see whether sentiment around this term had changed. From what I have observed, by the 18th century "virago" is a much less positive term. I expected, therefore, to see more negative sentiments over time. I decided to graph my results.

```python
def get_sentiments_data_by_decade(concordances, stopwords, skipWordNetPos=[]):
    score_by_decade = {}

    startYear = 1540 
    endYear = 1710 # ending in 1710 because I want to capture the texts from 1700
    numDecades = (endYear - startYear)/10 

    #modified version from the one used in class, instead of looping through document-by-document, loop through decade-by-decade
    for d in range(int(numDecades)):
        concordances_for_decade = get_concordances_by_year_range(startYear+d*10, startYear+(d+1)*10, concordances)
        tokens_for_decade = concordances_to_tokens(concordances_for_decade, stopwords)
        score, positives, negatives = get_sentiment_data_from_tokens(tokens_for_decade, skipWordNetPos)
        num_of_sentiment_words = len(positives) + len(negatives)

        if num_of_sentiment_words > 0:
            normalized_score = score / (len(positives) + len(negatives))
            score_by_decade[startYear+d*10] = normalized_score

    return score_by_decade
```

```python
sentiments_by_decade = get_sentiments_data_by_decade(virago_passages, virago_stopwords) 
```
