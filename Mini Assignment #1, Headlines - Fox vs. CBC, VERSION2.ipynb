```python
import urllib.request
response = urllib.request.urlopen('https://www.foxnews.com/')
FOXhtml = response.read()
```

```python
from bs4 import BeautifulSoup
FOXSoup = BeautifulSoup(FOXhtml, 'lxml')
```

```python
FOXHeadlines = []
FOXSoup = BeautifulSoup(FOXhtml, 'lxml')
FOXArticles = FOXSoup.find_all("article")
for FOXArticle in FOXArticles:
    FOXHeader = FOXArticle.find("header")
    if FOXHeader:
        FOXAnchor = FOXHeader.find("h2")
        FOXHeadlines.append(FOXAnchor.string)
```

```python
uniqueHeadlines = set(FOXHeadlines)
for uniqueHeadline in uniqueHeadlines:
    print(uniqueHeadline)
```

```python
CBCresponse = urllib.request.urlopen('https://www.cbc.ca/')
CBChtml = CBCresponse.read()
```

```python
CBCsoup = BeautifulSoup(CBChtml, 'lxml')
```

```python
CBCTitles = []
for CBCheadline in CBCsoup.find_all("h3", class_="headline"):
    CBCTitles.append(CBCheadline.text)
```

```python
CBCuniqueHeadlines = set(CBCTitles)
for CBCuniqueHeadline in CBCuniqueHeadlines:
    print(CBCuniqueHeadline)
```
